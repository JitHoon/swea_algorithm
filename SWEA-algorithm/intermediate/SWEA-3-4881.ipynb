## [파이썬 S/W 문제해결 기본] 5일차 - 배열 최소 합
### NxN 배열에 숫자가 들어있다. 한 줄에서 하나씩 N개의 숫자를 골라 합이 최소가 되도록 하려고 한다. 단, 세로로 같은 줄에서 두 개 이상의 숫자를 고를 수 없다. 조건에 맞게 숫자를 골랐을 때의 최소 합을 출력하는 프로그램을 만드시오.

### [입력] 첫 줄에 테스트 케이스 개수 T가 주어진다.  1≤T≤50
### 다음 줄부터 테스트 케이스의 첫 줄에 숫자 N이 주어지고, 이후 N개씩 N줄에 걸쳐 10보다 작은 자연수가 주어진다. 3≤N≤10

### [출력]
### 각 줄마다 "#T" (T는 테스트 케이스 번호)를 출력한 뒤, 합계를 출력한다.

#### `입력 3 | 3 | 2 1 2 5 8 5 7 2 2 ...`

#### `출력 #1 8 | #2 14 | #3 9`

```python
# dfs, 백트래킹

def find(row):
    global subsum, result

    # 백트래킹
    # 가장 작은 합을 찾아야하므로
    # subsum이 result보다 조금이라도 크면 탐색 종료
    # subsum이 작다면 row == N 이 될 때만 갱신

    if result < subsum:
        return

    # 마지막 row에 도달한 경우
    if row == N:
        # 기존 결과보다 작다면 result 갱신
        if subsum < result:
            result = subsum
        # 아니면 그냥 return
        return

    # 방문 여부를 열을 기준으로 반복 확인할 수 있다.
    # 방문하지 않은 열의 데이터를 subsum에 더하고 나서는 
    # 방문했다는 표시를 하고 다음 행으로 이동하는 방식이다.
    # 각 행에 대해 방문하지 않은 열의 데이터를 넣을 수 있다.
    for col in range(N):
        if not visit_col[col]:
            # 해당 col. 방문 후 subsum 증가
            visit_col[col] = True
            subsum += data[row][col]
            # 다음 row로 이동, 재귀적으로 마지막 row까지 호출
            print("재귀전", visit_col)
            print(subsum)

            find(row+1)
            print("재귀종료", visit_col)
            print(subsum, visit_col[col])

            # 다음 열에 대한 합이 전부 끝나면(row == N) 
            # result값을 갱신하고 return
            # 다른 열을 선택했을 때 subsum도 확인해야하므로 
            # 앞선 재귀가 끝나면 더해줬던 값을 다시 빼주고 방문 기록도 되돌려 놓는다.

            # if 문 밖에 이 소스코드를 적는다면
            # 방문한 모든 col 값을 subsum에 빼주게 되므로 #1 -22 를 출력한다.
            visit_col[col] = False
            subsum -= data[row][col]
            print("초기화", visit_col)
            print(subsum, visit_col[col])


for A in range(int(input())):
    N = int(input())
    data = [list(map(int, input().split())) for _ in range(N)]

    visit_col = [False]*N # 방문한 col 확인용

    subsum = 0
    result = 10*N+1 # Some big number

    find(0) # Start from row 0

    print('#{0} {1}'.format(A+1, result))
```
