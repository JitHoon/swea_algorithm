## [S/W 문제해결 기본] 5일차 - 계산기3
### 문자열로 이루어진 계산식이 주어질 때, 이 계산식을 후위 표기식으로 바꾸어 계산하는 프로그램을 작성하시오.
### 문자열 계산식을 구성하는 연산자는 +, * 두 종류이며 문자열 중간에 괄호가 들어갈 수 있다.
### 이 때 괄호의 유효성 여부는 항상 옳은 경우만 주어진다.
### 피연산자인 숫자는 0 ~ 9의 정수만 주어진다.

### [입력]
### 각 테스트 케이스의 첫 번째 줄에는 테스트 케이스의 길이가 주어진다. 그 다음 줄에 바로 테스트 케이스가 주어진다.
### 총 10개의 테스트 케이스가 주어진다.

### [출력]
### #부호와 함께 테스트 케이스의 번호를 출력하고, 공백 문자 후 답을 출력한다.

#### `입력 113 | 9+(5*2...`

#### `출력 #1 672676\n#2 1974171 ....`

```python
# stack 후위 표기식 정석
# ***
def postfix(data):
    stack = []
    output = []
    for i in data:
        if i in icp.keys():
            if not stack:
                stack.append(i)
            elif i == ')':
                while stack[-1] != '(':
                    output.append(stack.pop())
                stack.pop()
            else:
                while icp[i] <= isp[stack[-1]]:
                    output.append(stack.pop())
                stack.append(i)
        else:
            output.append(i)
    while stack:
        output.append(stack.pop())
    return output

def calculate(data):
    global result
    stack = []
    for i in data:
        if i == '+':
            result = stack.pop() + stack.pop()
            stack.append(result)
        elif i == '*':
            result = stack.pop() * stack.pop()
            stack.append(result)
        else:
            stack.append(int(i))
    return result

for test_case in range(1, 11):
    n = int(input())
    icp = {'(': 3, '*': 2, '+': 1, ')': 0}
    isp = {'(': 0, '*': 2, '+': 1}
    result = 0
    calculate(postfix(input()))
    print(f'#{test_case} {result}')
```
